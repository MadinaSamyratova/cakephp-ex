```python
import pandas as pd
from sklearn.datasets import load_wine

# LOADING THE WINE DATASET
wine = load_wine()
wine_data = pd.DataFrame(wine.data, columns = wine.feature_names)
wine_data['target'] = pd.Series(wine.target)

# SEARCHING FOR CATEGORICAL FEATURE
print("The unique value of column target are:")
print(wine_data['target'].unique())

# COMPUTING THE FREQUENCY
print("The frequency for each value of target:")
frequency = wine_data['target'].value_counts()/len(wine_data)
print(frequency)
```

# Task 2.
Load the file country-income.csv which includes numerical and categorical features. <br>
Perform data cleaning to replace any NaN values with the mean value of that particular feature. Then replace any categorical features with numerical features. Display the resulting dataset. <br>
You can use the sklearn.impute and sklearn.preprocessing packages to assist you. Include the code in your report.


**2. Answer** <br>
To complete this task, I imported SimpleImputer and Label Encoder from sklearn. <br>
Since there were operations to do on categorical and numerical features separately, I created subsets of the data set to only include one of the types at the time. <br>
Using the Single Imputer with specifying the strategy to be 'mean', I replaced the NaNs in the numerical data with the mean of the respective column. <br>
Then, using the Label Encoder, I reoplaced the categorical values with numerical ones. Each new value in the categorical features column would get a unique numeric value. <br>
Then, to get back to the original combined set, I concatinated the two data sets into one, resulting in the cleaned_country_income_data.

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

country_income_csv = pd.read_csv('country-income.csv')

num_feat = country_income_csv.select_dtypes(include=['number'])
cat_feat = country_income_csv.select_dtypes(include=['object'])

# REPLACING NANs IN NUMERICAL DATA SET WITH THE MEANS
num_imputer = SimpleImputer(strategy='mean')
num_feat = pd.DataFrame(num_imputer.fit_transform(num_feat), columns=num_feat.columns)
# TRANSFORMING CATEGORICAL FEATURES INTO NUMERICAL BY LABELLING
label_encoder = LabelEncoder()
cat_feat = cat_feat.apply(label_encoder.fit_transform)

cleaned_country_income_data = pd.concat([num_feat, cat_feat], axis=1)
cleaned_country_income_data
```

# Task 4.
Using the wine dataset from question 1, perform Principal Component Analysis (PCA) with 2 components. <br>
Transform the data and plot the scatterplot of all samples along the two principal components, color-coded according to the "target" column (this column is the class and should not be used in the PCA analysis). <br>


```python
from pandas._libs.tslibs.timedeltas import Components
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

wineData = load_wine()
wineDF = pd.DataFrame(wineData.data, columns = wineData.feature_names)
wineDF['target'] = pd.Series(wineData.target)

pca = PCA(n_components= 2)
principal_components = pca.fit_transform(wineDF.drop('target', axis=1))

principal_df = pd.DataFrame (principal_components, columns = ['pc1' , 'pc2'])
principal_df['target'] = wineData.target

colors = { 'class_0':'r', 'class_1':'g', 'class_2':'b'}

plt.figure(figsize=(6,5))
for target in wineData.target_names:
    d = principal_df[principal_df['target'] == wineData.target_names.tolist().index(target)]
    plt.scatter(d['pc1'], d['pc2'], c=colors[target], label=target, s=60)

plt.legend()
```

### Question
> What insights can you obtain by viewing the scatterplot of the principal components?
Can you easily distinguish the samples that belong to one class from the samples that belong to another class and so on? In other words, are the different classes (quite) distinctive one from the other, or is there a lot of overlap? If it is the latter, then why is this happening? <br>
### Answer
> By viewing the scatterplot of the principal components, it is possible to deduce how distinctive the classes are. As seen on the plot, none of the classes are perfectly distinct from each other, although the class 0 (red points) have the least overlap with the rest of the samples, whereas classes 1 and 2 (green and blue) are mixed together. <br>
The overlap of the two classes on the plot indicates that they share common characteristics or features in the lower-dimensional space created by PCA. This overlap is a result of PCA's dimensionality reduction, which may not fully capture the distinctions between these classes. <br>





### Question
>  What can be done to the data prior to performing PCA in order to alleviate this issue? <br>
 Do this action first and then perform PCA with 2 components, transform the data and plot the scatterplot of all samples along these two principal components, color-coded according to the "target" column.<br>
 Now are the different classes (quite) distinctive one from the other? Include the code in your report.
 ### Answer
> To alleviate the issue of the overlapping in the PCA plot, several preprocessing steps can be applied prior to performing the PCA. <br>
I decided to use normalisation, so that the features have the same scale and variance. Doing this ensured that no single feature dominates the PCA results. I used the z-score normalisation for this example by utilising the Standard Scaler.
<br>
All three classes are now distinct with only slight two overlaps.

```python
from sklearn.preprocessing import StandardScaler
# NORMALISATION
scaler = StandardScaler()
wineDF[wineData.feature_names] = scaler.fit_transform(wineDF[wineData.feature_names])

# NEW PCA PLOT
pca = PCA(n_components= 2)
principal_components = pca.fit_transform(wineDF.drop('target', axis=1))


principal_df = pd.DataFrame (principal_components, columns = ['pc1' , 'pc2'])
principal_df['target'] = wineData.target

colors = { 'class_0':'r', 'class_1':'g', 'class_2':'b'}

plt.figure(figsize=(6,5))
for target in wineData.target_names:
    d = principal_df[principal_df['target'] == wineData.target_names.tolist().index(target)]
    plt.scatter(d['pc1'], d['pc2'], c=colors[target], label=target, s=60)

plt.legend()
```

 Suppose that the class of the first observation is y1 = 1 and that the class of the second observation is y2 = 0. <br>
 How would a 1-nearest neighbour classifier based on the Euclidean distance classify the observation x3 = (3,2) and why?
 How would the same classifier classify the observation x4 = (0,1) and why? No code is needed for this task


### Answer
The given observations:
1. x1 = (1, 2) with class y1 = 1
2. x2 = (-1, 0) with class y2 = 0

For the observation x3 = (3, 2):
1. Euclidean distance between x3 and x1 = 2
2. Euclidean distance between x3 and x2 ≈ 4.47
Since the distance from x3 is larger to x2, the closest observation is x1 with the distance of 2. <br>
x1 has class y1 = 1, so x3 is classified as class 1.

For the observation x4 = (0, 1):
1. Euclidean distance between x4 and x1
≈ 1.41
2. Euclidean distance between x4 and x2 ≈ 1.41
The distance between x4 and x1 was calculated and found to be the same as the distance between x4 and x2. <br>
Since the distance from x4 to x1 and x2 is the same, I selected the class of the nearest neighbour by the order of them in the dataset - class 1 comes first in the data set, so x4 is classified as class 1. <br>
